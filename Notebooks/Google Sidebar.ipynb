```python
import pandas as pd
import time
import random
from bs4 import BeautifulSoup
import requests
import datefinder
```

# Google Sidebar

Google provides airdates (but no other information) in the sidebar when searching for an episode. However, I cannot find any source for this information, and it does not appear to come from any of the top linkes like Official Family Feud sources, IMDB, Trakt, TV Guide, etc. 

Thus, we Google search each episodes and extract the airdates into a dataframe. This will be especially helpful for Season 23/24, where TV Guide/Trakt have limited date/family information available. While, Bobby does has family info for these seasons, his episodes are only labled by date and not by season/episode number.

```python
def get_google_airdate(season, episode):
    """ Use Google sidebar information to retrieve airdate of episode from provided season/episode number"""
    
    # Print season/episode number and sleep between runs
    print(item)
    time.sleep(5*random.random())
    
    # Parse Google search results for single episode to soup
    url = f'https://www.google.com/search?q=Family+Feud+episode+(season+{season},+episode+{episode})'
    soup = BeautifulSoup(requests.get(url).text, 'html.parser')

    # The dates we want have the month spelled out, so use this to narrow the search (see example photo)
    # Only include spans that have a month name in them
    months = ["January", "February", "March", "April", "May", "June", "July", "August", "September", "October", "November", "December"]
    
    # Based on my preliminary search, the sidebar appears above the other search results within the soup
    # Thus, we can extract the first date that appears and this should be the airdate
    # Still requires a manual check since scattered episodes are missing a date and then a wrong date is filled in
    
    # Get list of all spans on page if that span includes a written-out month
    span_list = []
    spans = soup.find_all('span')
    for span in spans:
        if any(month in span.text for month in months):
            span_list.append(span.text)
    
    def get_first_date_from_list(lis):
        """ Parse the first date from a list of strings"""
        for item in lis:
            matches = datefinder.find_dates(item)
            try:
                first_date = next(matches)
                return first_date
            except:
                pass
        return False
    
    # Get first date from span list
    first_date = get_first_date_from_list(span_list)
    
    return first_date
```

```python
df_all = pd.read_pickle("../pickles/df_all.pkl")
ep_list = df_all.sort_values(by=['SeasonNum','EpisodeNum'], ascending=False).to_dict('records')
date_list = []
for item in ep_list:
    ep = [item['SeasonNum'], item['EpisodeNum'], get_google_airdate(item['SeasonNum'],item['EpisodeNum'])]
    date_list.append(ep)
df_google = pd.DataFrame(date_list, columns=['SeasonNum','EpisodeNum','AirDate_Google'])
df_google.to_pickle("../pickles/df_google.pkl")
```
